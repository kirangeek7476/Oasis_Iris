# Simplest Model

## This is Very much to decision tree algorithm

```python
def single_feature_prediction(petal_length):
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
df_train.columns
```

```python
x_train
```

```python
x_train[:,2]
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
manual_y_predictions
```

## Manual Model Accuracy

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f"Manual Model accuracy is : {manual_model_accuracy * 100:.4f}%")
```

## Where are we misclassifying

```python
y_pred = cross_val_predict(model,x_train,y_train,cv=5)
```

```python
predicted_correctly = (y_pred == y_train)
```

```python
x_train[predicted_correctly]
```

```python
not_predicted_correctly = ~predicted_correctly
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

## Comparing with the scatterplot  using sepal width

```python
sns.scatterplot("sepal length (cm)","sepal width (cm)",hue = "prediction_label",data = df_predictions)
```

```python
sns.scatterplot("sepal length (cm)","sepal width (cm)",hue = "species_names",data = df_predictions)
```

```python
def plot_incorrect(df_predictions,x_axis_feature,y_axis_feature):
    flg,axs = plt.subplots(2, 2, figsize = (10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature, hue="prediction_label",data=df_predictions,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature, hue="species_names",data=df_predictions,ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature, hue="correct_predictions",data=df_predictions,ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect(df_predictions, "sepal length (cm)","sepal width (cm)")
```

# Model Tuning

```python
for reg_param in (0.1,0.3,0.9,1,1.3,1.9,2):
    model = LogisticRegression(max_iter = 200,C = reg_param)
    accuracies = cross_val_score(model,x_train,y_train,cv = 5,scoring = "accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
```
